```python
import json
from azure.storage.blob.aio import ContainerClient
from azure.storage.blob.aio import BlobClient
```

```python
with open("../pylanche/config.json", "r") as config_file:
    config = json.load(config_file)
    BLOB_STORAGE_CONNECTION_STRING = config['BLOB_STORAGE_CONNECTION_STRING']
```

```python
try:
    container_client = ContainerClient.from_connection_string(conn_str=BLOB_STORAGE_CONNECTION_STRING, container_name="pylanche-state")
    await container_client.create_container()
except Exception as error:
    print(error)
```

```python
try:
    blob = BlobClient.from_connection_string(conn_str=BLOB_STORAGE_CONNECTION_STRING, container_name="pylanche-state", blob_name="state")
    with open("../pylanche/state.db", "rb") as data:
        await blob.upload_blob(data)
except Exception as error:
    print(error)
```

```python
try:
    blob = BlobClient.from_connection_string(conn_str=BLOB_STORAGE_CONNECTION_STRING, container_name="pylanche-state", blob_name="state")
    with open("../pylanche/state.db", "wb") as state_blob:
        stream = await blob.download_blob()
        data = await stream.readall()
        state_blob.write(data)
except Exception as error:
    print(error)
```
