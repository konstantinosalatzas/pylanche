```python
import sqlite3
```

```python
# Connect to db.
con = sqlite3.connect("../pylanche/state.db")
cur = con.cursor()
```

```python
def select_all_from_state(table: str) -> list:
    res = cur.execute("SELECT * FROM {}".format(table))
    return res.fetchall()
```

```python
print(select_all_from_state("state"))
```
